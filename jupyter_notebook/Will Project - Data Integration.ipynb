### Data Integration

```python
# Code to import libraries as you need in this assessment
import numpy as np
import pandas as pd
import datetime
import re
import random
```

```python
df1 = pd.read_csv('../cleaned_csv/cleaned_gov_measured.csv')
df2 = pd.read_csv('../cleaned_csv/covid19_cumulative_monthly_stats.csv')
df3 = pd.read_csv('../cleaned_csv/covid19_monthly_statistics.csv')
# Display two dataframes together to compare
print('S1:' + str(df1.shape))
display(df1.head())
print('S2:' + str(df2.shape))
display(df2.head())
print('S3:' + str(df3.shape))
display(df3.head())
```

```python
#Rename the STEPS_TAKEN column to month_year gov_measures_subset
df1.rename(columns = {'DATE_IMPLEMENTED':'month_year'}, inplace = True)
```

```python
print (df1.info())
print (df2.info())
print (df3.info())
```

```python
df = pd.merge(df2, df3, on=['ISO','REGION','COUNTRY','month_year','population'])
df.head()
```

```python
df.info()
```

```python
df_final = pd.merge(df1, df, on=['ISO','REGION','COUNTRY','month_year'])
df_final.head()
```

```python
df_final.info()
```

```python
df_final.drop(['ISO','REGION'], inplace=True, axis=1)
```

```python
df_final.info()
```

```python
### check duplication
duplicates = df_final[df_final.duplicated(keep=False)] # showing all duplicated records
print ("There are "+ str(len(duplicates)) + " duplicate records found")
duplicates.sort_values(by=duplicates.columns.tolist()).head()
```

```python
df_final = df_final.drop_duplicates(keep='last')
df_final.shape
```

```python
df_final.info()
```

```python
df_final['COUNTRY'].unique()
```

```python
# Writing the clean dataframe to csv file
df_final.to_csv('../cleaned_csv/integrated_data.csv', index=False)
```
