```python
def updatecol(monthyear, prev_monthyear, colname, newcol):
    #prev = month_order[month_order.index(month_year)-1]
    #prevmonthcases = data.loc[data.month_year==prev].iloc[0,16]
    for c in countries:
        if(data.month_year):
            prevmonthcases = list(data.loc[(data.month_year==prev_monthyear) & (data.country==c), colname])
            newcases = list(data.loc[(data.month_year==monthyear) & (data.country==c), colname])
            #print(c)
            if newcases and prevmonthcases:
                if prevmonthcases[0]!=0:
                    increase = (newcases[0]-prevmonthcases[0])*100/prevmonthcases[0]
                else:
                    increase = 0
                data.loc[(data.month_year==monthyear) & (data.country==c),newcol] = increase
            else:
                data.loc[(data.month_year==monthyear) & (data.country==c),"newcol"] = 0
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=data.loc[:,["population","stringency_index", "total_cases", "total_tests", "new_cases", "new_deaths"]]
target = data.loc[:,"social distancing"]
```

```python
new_ind = [] # current state
cur_max_score = 0.0 
col_num = dt.shape[1] # number of features/columns
col_ind_random = shuffle(range(col_num), random_state=1)

for cur_f in range(col_num):
    new_ind.append(col_ind_random[cur_f])
    new_data = dt.iloc[:, new_ind]
    X_train, X_test, y_train, y_test = train_test_split(new_data, target, test_size=0.4, random_state=0)
    clf = DecisionTreeClassifier() #KNeighborsClassifier(5, weights='distance', p=1)
    fit = clf.fit(X_train, y_train)
    cur_score = fit.score(X_test, y_test)
    print(f"Score with {len(new_ind)} features: {cur_score:.3f}")
    if cur_score < cur_max_score:
        new_ind.remove(col_ind_random[cur_f])
    else:
        cur_max_score = cur_score
        print(f" --- Score with {len(new_ind)} features: {cur_score:.3f}  --- ")
print(f'\n The best subset of features (so far) is \n{new_ind}')
#print(f'\n The best subset of features (so far) by name is \n{dt.feature_names[new_ind]}')
print(f'\n The order of the features: \n{col_ind_random}')
```
